```python
from plotly import tools, offline
from plotly.graph_objs import *
from plotly.offline import init_notebook_mode, iplot, iplot_mpl
import plotly.express as px
import plotly.graph_objs as go


import pandas as pd
```

```python
df_stat = pd.read_csv("trading_statistic.csv")
df_stat['date'] = pd.to_datetime(df_stat['date'])
df_stat
```

```python
buy = df_stat[(df_stat['action']==1) & (df_stat['action'].shift(1) == 0)]
sell = df_stat[(df_stat['action']==0) & (df_stat['action'].shift(1) == 1)]
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x= df_stat['date'], y=df_stat['portfolio_valuation'], name = "баланс"))
fig.add_trace(go.Scatter(x=buy['date'], y=buy['portfolio_valuation'], mode='markers', marker_symbol='triangle-up', name='buy', marker_size=20, line=dict(color='green',width=8)))
fig.add_trace(go.Scatter(x=sell['date'], y=sell['portfolio_valuation'], mode='markers', marker_symbol='triangle-down', name='sale', marker_size=20, line=dict(color='red',width=8)))
fig.update_layout(
        title="График",
        xaxis_title="Дата торговли",
        yaxis_title="Количество денег"
        #yaxis_title="Баланс после сделки"
)
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x= df_stat['date'], y=df_stat['last_price'], name = "цена"))
fig.add_trace(go.Scatter(x=buy['date'], y=buy['last_price'], mode='markers', marker_symbol='triangle-up', name='buy', marker_size=20, line=dict(color='green',width=8)))
fig.add_trace(go.Scatter(x=sell['date'], y=sell['last_price'], mode='markers', marker_symbol='triangle-down', name='sale', marker_size=20, line=dict(color='red',width=8)))

fig.update_layout(
        title="График",
        xaxis_title="Дата торговли",
        yaxis_title="цена"
        #yaxis_title="Баланс после сделки"
)
fig.show()
```
